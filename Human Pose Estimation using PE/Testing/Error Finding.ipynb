```python
path =gb.glob('/content/Google_Drive/MyDrive/College /Human Pose Estimation Using Open CV/Images/Estimated Coordinates/*.csv')
```

```python
def distance(x1,y1,n1,x2,y2,n2):
      err=[]
      if(len(x1) > len(x2)):
           x=len(x2)
      else:
           x=len(x1)      
      for i in range(x):
            nv=(float)(mt.sqrt( ((x1[i]-x2[i])**2)+((y1[i]-y2[i])**2)))
            err.append(nv)
      return err
```

```python
fields = ['Joint Name','Est. Coordinates X','Est. Coordinates Y','Joint Name','Act. Coordinates X','Act. Coordinates Y','Approx Error']
for file in path:
      path1= file.replace('Estimated Coordinates','Actual Coordinates')
      path2= file.replace('Estimated Coordinates','Error')
      f1=[]
      f2=[]
      f3=[]
      f4=[]
      f5=[]
      f6=[]
      print('Path of Estimated Coordinates File-',file)
      print('Path of Actual Coordinates File- ',path1)
      print('Path of Error File - ',path2)
      take=pd.read_csv(file)
      var1=take['Coordinates X']
      var2=take['Coordinates Y']
      var3=take['Joint Name']
      take2=pd.read_csv(path1)
      var4=take2['X']
      var5=take2['Y']
      var6=take2['class']
      #f1 - f6 all are list fot taking different values of both file in a list
      f1.extend(var1) # Est X
      f2.extend(var2) # Est Y
      f3.extend(var3) 
      f4.extend(var4) # Act X
      f5.extend(var5) # Act Y
      f6.extend(var6)
      newfile=distance(f1,f2,f3,f4,f5,f6)
      with open(path2, 'w') as csvfile: 
        # creating a csv writer object 
        csvwriter = csv.writer(csvfile) 
         
        # writing the fields 
        csvwriter.writerow(fields) 
            
        # writing the data rows 
        for i in range(len(f1)):
            csvwriter.writerow([f3[i],f1[i],f2[i],f6[i],f4[i],f5[i],newfile[i]])
      del f1
      del f2
      del f3
      del f4
      del f5
      del f6 
```
